# Consumer resource model

$J_{E. coli} = \mu_{E. coli}\frac{R}{R + K_{M}}$   
$J_{Salmonella} = \mu_{Salmonella}\frac{R}{R + K_{M}}$

$J$: per capita growth rate  
$\mu$: maximum growth rate  
R: limitting resource concentration  
$K_{M}$: Monod constance

```python
exp = Experiment()
exp.isoclines()
```

$\frac{dSalmonella}{dt}=J_{Salmonella}*Salmonella$  

$\frac{dE. coli}{dt}=J_{E. coli}*E. coli$

$\frac{dR}{dt} = R - J_{Salmonella} * \frac{Salmonella}{Y_{Salmonella}} - J_{E. coli} * \frac{E. coli}{Y_{E. coli}}$

$Y$: Yield

```python
exp = Experiment()

exp.dilution_factor = 100
exp.total_transfers = 4
exp.simulate_experiment("consumer_resource")
exp.plot_N()
exp.plot_R()
```

# Susceptibility of E. coli to Cefotaxime

$\frac{dSalmonella}{dt}=\mu_{E. coli}\frac{R}{R + K_{M}} - \omega\frac{T}{T + K_{T}}$

$\omega$: maximum death rate  
$T$: Antibiotic
$K_T$: Monod constance

```python
exp = Experiment()
exp.dilution_factor = 1
exp.total_transfers = 1
exp.simulate_experiment("e_toxin")
exp.plot_e_coli()
```

# Susceptibility of Salmonella to Chloramphenicol

$\frac{dSalmonella}{dt}=\mu_{E. coli}\frac{R}{R + K_{M}}\frac{K_T}{K_T + T}$

$\omega$: maximum death rate  
$T$: Antibiotic
$K_T$: Monod constance

```python
exp = Experiment()
exp.dilution_factor = 1
exp.total_transfers = 1
exp.simulate_experiment("s_toxin")
exp.plot_salmonella()
```

# Salmonella protects E. coli

$J_{Salmonella} = R \frac{R}{R + K_M}(1 - f)$  

$\frac{dCefotaxime}{dt} = -Cefotaxime * (f * a * J_{Salmonella}) *Salmonella$

$f$: energy investment  
$a$: degradation rate

$f$: energy investment  
$a$: degradation rate
