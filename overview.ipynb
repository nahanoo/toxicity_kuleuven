# Competition for a single limitting resource

```python
exp = Experiment()
Rs = np.linspace(0,30)
J_e_coli = [exp.e.r * R / (R + exp.e.Km) for R in Rs]
J_salmonella = [exp.s.r * R / (R + exp.s.Km) for R in Rs]
plt.plot(Rs,J_e_coli,label='E_coli')
plt.plot(Rs,J_salmonella,label='Salmonella')
plt.xlabel('Resource [mM]')
plt.ylabel('Per capita growth rate [1/h]')
plt.legend()
plt.show()

exp.dilution_factor = 100
exp.total_transfers = 8
exp.simulate_experiment("consumer_resource")
exp.plot_N()
```
